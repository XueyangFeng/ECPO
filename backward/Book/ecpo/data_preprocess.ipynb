```python
import json

def read_jsonl_file(file_path):
    data_list = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            try:
                data = json.loads(line)
                data_list.append(data)
            except json.JSONDecodeError as e:
                print(f"解析错误：{e}, 文件: {file_path}, 行内容: {line[:50]}")
    return data_list

# 从文件中加载 JSON 数据
def load_json(file_path: str):
    with open(file_path, 'r', encoding='utf-8') as file:
        data = json.load(file)
    return data
# 打开并读取 JSONL 文件
file_path = './amazon_book_path.jsonl'  # 请替换为你的 JSONL 文件路径

path_list = read_jsonl_file(file_path)

result_list = []  # 用来存储所有的 input-output 对象

for data in path_list:
    try:
        scratchpad = data["scratchpad"]
        
        # 遍历 scratchpad 中的每个键值对
        for key, actions in scratchpad.items():
            print(actions)
            # 遍历每个 action
            for action_key, action_data in actions.items():
                if action_key.startswith('action '):
                    print(action_data)
                    input_data = action_data.get('input', 'No input available')
                    output_data = action_data.get('output', 'No output available')
                    
                    # 将 input 和 output 组成一个字典并添加到结果列表
                    result_list.append({
                        "system": "You are a helpful assistant",
                        "Instruction": input_data,
                        "input": "",
                        "output": output_data,
                        "rec_reward": actions["rec_reward"],
                        "action_reward": actions["action_reward"],
                        "exp_reward": actions["exp_reward"],               
                        "action_reason": actions["action_reason"],
                        "exp_reason": actions["exp_reason"],     
                        "action_reason": actions["action_reason"],         
                    })
                
    except json.JSONDecodeError as e:
        print(f"解析错误：{e}")

# 输出最终的结果列表
#print(json.dumps(result_list, indent=4))  # 美化输出，查看整个列表

# 可选择保存到一个新的 JSON 文件
output_file = './amazon_book_ecpo_raw.json'  # 请替换为你想保存的路径
with open(output_file, 'w', encoding='utf-8') as outfile:
    json.dump(result_list, outfile, ensure_ascii=False, indent=4)
```
